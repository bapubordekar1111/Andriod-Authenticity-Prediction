```python
# Splitting original data set for imputing values for related apps
df_train, df_test = datawig.utils.random_split(df)
```

```python
# inputting required for algorithm
imputer = datawig.SimpleImputer(
    input_columns=['Category','Related apps'], # column(s) containing information about the column we want to impute
    output_column= 'Related apps', # the column we'd like to impute values for
    output_path = 'imputer_model' # stores model data and metrics
    )
```

```python
# applying algorithm
imputer.fit(train_df=df_train, num_epochs=50)
```

```python
# to convert number of ratings into sub groups
def rat_to_groups(no_of_ratings):
  '''
  function to convert ratings to resp translation
  '''
  if no_of_ratings <= 10:
    return 'ratings between 0-10'
  elif (no_of_ratings >= 10) and (no_of_ratings <1000 ):
    return 'ratings between 10 and 1000'
  else:
    return 'ratings greater than 1000'
```

```python
#plot for no of user ratings each app
df['Rating_cate_dummy'] = df['Number of ratings'].apply(lambda x: rat_to_groups(x) )
plt.figure(figsize=(15,9))
ax = sns.countplot(x="Rating_cate_dummy", data= df, palette='husl')
plt.xlabel('Sub groups ')
plt.ylabel('Number of applications')
plt.title('Number of users ratings each app')
```

# Text Column: Preparing for NLP Modelling

## 1. Removing HTML Tags on Description Column

```python
#remove html tag
import re
def remove_tags(string):
    result = re.sub('<.*?>','',str(string))
    return result
df['Description'] = df['Description'].apply(lambda cw : remove_tags(cw))
```

## 2. Cleaning Package Column

```python
#join column
def join_package(string):
  m=string.split('.')
  return ' '.join([word for word in m])
df['Package'] = df['Package'].apply(join_package)
```

## 3. Cleaning All columns from other characters

```python
#remove other chara
import re
def cleaned(x):
    return re.sub(r"[^a-zA-Z ]","",str(x))
df['Description'] = df['Description'].apply(cleaned)
df['App'] = df['App'].apply(cleaned)
df['Package'] = df['Package'].apply(cleaned)
```

## 4. All char to lower

```python
#all chara to lower
df['Description'] = df['Description'].str.lower()
df['App'] = df['App'].str.lower()
df['Package'] = df['Package'].str.lower()
df['App']
```

## 5. Combining text columns to single column 

```python
#combine all text to single
df['Text_combined'] = df['Description'] + ' ' + df['Package'] + ' ' + df['App'] 
```

```python
df['Text_combined'].iloc[0]
```

## 6. Stop Words, Tokenisation and Stemming! 

```python
#stop words, tokenisation and stemming import
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem.snowball import SnowballStemmer
nltk.download('punkt')
stemmer = SnowballStemmer('english')
stop_words = set(stopwords.words('english'))
#com and android seems to be a recurring word
stop_words.update({'com','android'})
```

```python
#function for filtering and applying
def wordfilter(string, filtwords):
    filtered = []
    tokens = word_tokenize(string) 
    for word in tokens:
        if word not in filtwords:
            filtered.append(stemmer.stem(word))
    return filtered

df['Filtered'] = ''
for item, row in df.iterrows():
    df.at[item, 'Filtered'] = wordfilter(row['Text_combined'], stop_words)
def join_words(x):
  return " ".join(x)

df['Filtered'] = df['Filtered'].apply(join_words)
    
df['Filtered']
```

```python
# Models from Scikit-Learn
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
from xgboost import XGBClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC, SVR
```

```python
# Model Evaluations
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.model_selection import RandomizedSearchCV, GridSearchCV
from scipy.stats import uniform, randint
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.metrics import precision_score, recall_score, f1_score
from sklearn.metrics import plot_roc_curve
```

```python
# Put models in a dictionary
models = {"Logistic Regression": LogisticRegression(max_iter=10000),
          "Random Forest": RandomForestClassifier(),
          "Gradient Boosting": GradientBoostingClassifier(),
          "XGBoosting": XGBClassifier(),
          "KNN":KNeighborsClassifier()}

# Create a function to fit and score models
def fit_and_score(models, X_train, X_test, y_train, y_test):
    """
    Fits and evaluates given machine learning models.
    models : a dict of differetn Scikit-Learn machine learning models
    X_train : training data (no labels)
    X_test : testing data (no labels)
    y_train : training labels
    y_test : test labels
    """
    # Set random seed
    np.random.seed(42)
    # Make a dictionary to keep model scores
    model_scores = {}
    # Loop through models
    for name, model in models.items():
        # Fit the model to the data
        model.fit(X_train, y_train)
        # Evaluate the model and append its score to model_scores
        model_scores[name] = model.score(X_test, y_test)
    return model_scores
```

```python
model_scores = fit_and_score(models=models,
                             X_train=X_train,
                             X_test=X_test,
                             y_train=y_train,
                             y_test=y_test)

model_scores
```
